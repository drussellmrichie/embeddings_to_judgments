```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
domains = ['trait', 'risk', 'people', 'food', 'occupation', 'brand', 'good']
```

```python
results_files  = ['traits.csv',
                 'risks.csv',
                 'people.csv',
                 'foods.csv',
                 'occupations.csv',
                 'brands.csv',
                 'goods.csv']
```

```python
# results_files = [x for x in os.listdir('results/dictionary_results') if 'csv' in x]
# results_files
```

```python
fig, ax_array = plt.subplots(nrows=4, ncols=2, figsize=(20,25))

for ax, results_file in zip(fig.axes, results_files):
    df = pd.read_csv(f'results/dictionary_results/{results_file}')
        
    sns.barplot(y='construct', x='difference_z', data=df, color='gray', ax=ax)
    ax.set_ylabel('')
    ax.set_xlim(-3,3)
    dim1 = df.columns[2].capitalize()
    dim2 = df.columns[3].capitalize()
    
    ax.set_xlabel(f'{dim1} - {dim2} z-score', size=22)
    ax.tick_params(axis='y', labelsize=20)
    ax.tick_params(axis='x', labelsize=20)
    
ax = fig.axes[-1]
fig.delaxes(ax)
fig.tight_layout()
plt.savefig('figures/dictionary_zscores_barplots.eps', format='eps')
plt.savefig('figures/dictionary_zscores_barplots.png', format='png')
plt.savefig('figures/dictionary_zscores_barplots.svg', format='svg')
```
