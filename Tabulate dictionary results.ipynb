```python
import pandas as pd
```

```python
dims = [('masculine', 'feminine'),
       ('dread-inducing', 'unknowable'),
       ('warm', 'competent'),
       ('tasty', 'nutritious'),
       ('significance', 'autonomy'),
       ('sincere', 'exciting'),
       ('hedonic', 'utilitarian')]

flattened_dims = [dim for pair in dims for dim in pair]
```

```python
all_dims_df = pd.concat([pd.read_csv(f'results/predicted_judgment_associates/{dim}_dictionary_words.csv', 
                                     header=None, 
                                     names=[dim, f'pred_{dim}_association']) for dim in flattened_dims], 
                        axis=1)
all_dims_df.head()
```
